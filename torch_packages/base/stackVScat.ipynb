```python
import torch

base = []

for _ in range(5):
    tmp = torch.randn(5,4)
    base += [tmp]
```

```python
base
```

```python
torch.stack(base) 

# stack은 list안에 있는 element들을 맨앞의 차원을 추가하고 element수 만큼 많아짐.
```

```python
torch.cat(base, dim = 0)

# cat은 list안에 있는 element의 dim에다가 붙여 더한다.
# cat을 쓰면 안에 잇는 element.shape(*dim) == torch.cat 결과물
```
